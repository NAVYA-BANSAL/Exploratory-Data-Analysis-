#### Chart - 3

````python
# Chart - 3 visualization code
#Guage Chart - Numeric Column Analysis
#Chart for Changed Credit Limit

avg_limit_change = data['Changed_Credit_Limit'].mean()

fig = go.Figure(go.Indicator(
    mode="gauge+number",
    value=avg_limit_change,
    title={
        'text': "Average Changed Credit Limit"}, gauge={
            'axis': {
                'range': [data['Changed_Credit_Limit'].min(), data['Changed_Credit_Limit'].max()
                ]
            },
            'bar': {'color': 'black'},
            'steps': [
                {
                    'range': [data['Changed_Credit_Limit'].min(), data['Changed_Credit_Limit'].quantile(0.33)], 'color': 'lightcoral'
                },

                {
                    'range': [data['Changed_Credit_Limit'].quantile(0.33), data['Changed_Credit_Limit'].quantile(0.66)], 'color': 'khaki'
                },

                {
                    'range': [data['Changed_Credit_Limit'].quantile(0.66), data['Changed_Credit_Limit'].max()], 'color': 'lightgreen'
                }
            ]
    }
))

fig.update_layout(title='Gauge Chart for Average Changed Credit Limit')
fig.show()
````

##### 1. Why did you pick the specific chart?

Answer Here.

A gauge chart is effective for displaying a single key performance metric against predefined ranges. It provides a quick visual indication of whether the average changed credit limit falls within a low, medium, or high range.

##### 2. What is/are the insight(s) found from the chart?

Answer Here

The average changed credit limit is approximately 10.47, which falls within the medium range. This suggests that, on average, customers experience moderate changes in their credit limits rather than extremely low or high adjustments.
